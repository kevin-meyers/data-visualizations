```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('/home/kevin/Downloads/AppleStore.csv')
```

```python
df = df.drop(['currency', 'id'], axis=1)
```

```python
sns.heatmap(df[['user_rating', 'price']].corr())
```

```python
sns.jointplot(x='user_rating', y='price', data=df.loc[df['price'] < 6], kind='kde')
```

```python
cont_rating_dict = {'4+': 0, '9+': 1, '12+': 2, '17+': 3}
```

```python
df['cont_rating'] = df['cont_rating'].apply(lambda rating: cont_rating_dict[rating])
```

```python
sns.jointplot(data=df, x='cont_rating', y='user_rating', kind='kde')
```

```python
df['size_bytes'] = df['size_bytes'].apply(lambda size: np.log2(size))
```

```python
sns.jointplot(x='size_bytes', y='user_rating', data=df, kind='kde')
```
