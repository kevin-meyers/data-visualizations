```python
def z_conf_int(sample, significant_level, sigma):
    num_samples = len(sample)
    z_score = scipy.stats.norm.ppf(1 - significant_level/2)
    X_bar = np.mean(sample)
    L = X_bar - z_score*sigma/np.sqrt(num_samples)
    U = X_bar + z_score*sigma/np.sqrt(num_samples)
    
    return L, U
```
